Оскільки в даних є "ЮУАЕС" та "ПАЕС" (це одна і та сама Південноукраїнська АЕС, яку перейменували), необхідно їх об'єднати в одну ПАЕС. Це допоможе уникнути помилок, коли при аналізі показників для кожної із станції.

```python
df['station'] = df['station'].replace('ЮУАЕС', 'ПАЕС')
print(f"Список станцій: {df['station'].unique()}")
```

```python
rows_with_nan = df[df.isnull().any(axis=1)]
display(rows_with_nan)
```

Аналіз пропущених значень виявив два ключові аспекти. По-перше, з 2022 року повністю відсутні дані щодо Запорізької АЕС, що пов’язано з воєнними діями. По-друге, для Хмельницької АЕС відсутні дані за показником co_60_dump (скиди Кобальту-60). Це може свідчити про специфіку моніторингу або про незафіксовані нульові значення. 
Ці особливості необхідно враховувати під час подальшого аналізу та візуалізації результатів. Видалення таких записів є недоцільним через невеликий обсяг даних. Проте заповнення пропусків середніми значеннями також не є коректним для даних екологічного та радіологічного характеру, оскільки це може спотворити реальну картину показників.
